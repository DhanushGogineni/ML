## Import libraries

### Subtask:
Import all necessary libraries for data manipulation and visualization (numpy, pandas, matplotlib, and seaborn).


```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
data = pd.read_csv("2020-2025.csv")
data.head()
```

## Categorical Data Visualization

Let's visualize the data for a few selected countries to demonstrate the categorical plots.

```python
# Question: Which selected country had the highest value in 2023?
# Answer: Based on the bar graph, Algeria had the highest value in 2023.

selected_countries = ['Algeria', 'Angola', 'Albania', 'Andorra']
categorical_data = data[data['Country'].isin(selected_countries)]

plt.figure(figsize=(7, 6))
plt.bar(categorical_data['Country'], categorical_data['2023'], width=0.4)
plt.title('2023 Data for Selected Countries')
plt.xlabel('Country')
plt.ylabel('Value')
plt.show()
```

```python
plt.figure(figsize=(7, 6))
sns.countplot(x='Country', data=categorical_data, order=selected_countries, width=0.4)
plt.title('Count of Selected Countries')
plt.xlabel('Count')
plt.ylabel('Country')
plt.show()
```

## Pie Chart

Now, let's create a pie chart to visualize the proportion of a selected year's data for the chosen countries.

```python
plt.figure(figsize=(6, 6))
plt.pie(categorical_data['2023'], labels=categorical_data['Country'], autopct='%1.1f%%', startangle=140)
plt.title('Proportion of 2023 Data for Selected Countries (Pie Chart)')
plt.axis('equal') # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()

# Question: What proportion of the 2023 data do Algeria and Angola represent combined for the selected countries?
# Answer: Based on the pie chart, Algeria represents 64.4% and Angola represents 28.5%, totaling 92.9% of the 2023 data for the selected countries.
```

## Continuous Data Visualization

Let's visualize the distribution of a continuous variable using a histogram, KDE plot, and box plot.

```python
continuous_data = data['2023'].dropna()

# Histogram
plt.figure(figsize=(7, 6))
plt.hist(continuous_data, bins=20, edgecolor='black')
plt.title('Distribution of 2023 Data (Histogram)')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()

# Question: What is the approximate range of the most frequent values in the 2023 data distribution?
# Answer: Based on the histogram, the most frequent values in the 2023 data distribution are approximately in the range of 0 to 50,000.
```

```python
plt.figure(figsize=(10, 6))
sns.kdeplot(continuous_data, fill=True)
plt.title('Distribution of 2024 Data (KDE Plot)')
plt.xlabel('Value')
plt.ylabel('Density')
plt.show()

# Question: Where does the KDE plot indicate the highest density of 2023 data values?
# Answer: The KDE plot indicates the highest density of 2023 data values is close to 0.
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x=continuous_data)
plt.title('Distribution of 2023 Data (Box Plot)')
plt.xlabel('Value')
plt.show()

# Question: What does the box in the box plot represent?
# Answer: The box in the box plot represents the interquartile range (IQR), which contains the middle 50% of the data.
```
